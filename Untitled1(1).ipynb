```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('creditcard.csv')
```

```python
##Data Exploration
print(df.info())
print(df['Class'].value_counts())
print(df.describe())
```

```python
##Visualize Class Imbalance
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: Legit, 1: Fraud)')
plt.show()
```

```python
##Explore Transaction Amounts
plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==0]['Amount'], bins=50, color='blue', label='Legit', kde=True)
sns.histplot(df[df['Class']==1]['Amount'], bins=50, color='red', label='Fraud', kde=True)
plt.legend()
plt.title('Transaction Amounts by Class')
plt.xlabel('Amount')
plt.show()
```

```python
##Correlation Heatmap
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
## time based analysis
df['Hour'] = df['Time'].apply(lambda x: np.floor(x / 3600) % 24)
plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==1]['Hour'], bins=24, color='red', label='Fraud', kde=False)
sns.histplot(df[df['Class']==0]['Hour'], bins=24, color='blue', label='Legit', kde=False, alpha=0.5)
plt.legend()
plt.title('Transaction Hour Distribution')
plt.xlabel('Hour of Day')
plt.show()
```

```python
##Feature Distribution Comparison
features = ['V1', 'V10', 'V14']  # Known to correlate with fraud
for feature in features:
    plt.figure(figsize=(8,4))
    sns.kdeplot(df[df['Class']==0][feature], label='Legit', fill=True)
    sns.kdeplot(df[df['Class']==1][feature], label='Fraud', fill=True, color='red')
    plt.title(f'Distribution of {feature} by Class')
    plt.legend()
    plt.show()
```

```python
##Boxplot of Amounts
plt.figure(figsize=(8,6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amounts by Class')
plt.show()
```

```python
fraud_ratio = df['Class'].mean()
print(f"Fraudulent transactions: {fraud_ratio:.4%} of all transactions")
```

```python
##Cumulative Fraud Loss
total_fraud = df[df['Class']==1]['Amount'].sum()
print(f"Total fraud loss in dataset: ${total_fraud:,.2f}")

# Cumulative loss over time
df_fraud = df[df['Class']==1].sort_values('Time')
df_fraud['Cumulative_Loss'] = df_fraud['Amount'].cumsum()
plt.figure(figsize=(10,5))
plt.plot(df_fraud['Time'], df_fraud['Cumulative_Loss'])
plt.title('Cumulative Fraud Loss Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Cumulative Loss ($)')
plt.show()
```

```python
## Hourly Fraud Rate
df['Hour'] = df['Time'].apply(lambda x: np.floor(x / 3600) % 24)
hourly_fraud = df.groupby('Hour')['Class'].mean()
plt.figure(figsize=(10,5))
hourly_fraud.plot(kind='bar', color='red')
plt.title('Fraud Rate by Hour of Day')
plt.ylabel('Fraud Rate')
plt.xlabel('Hour')
plt.show()
```



    

    Fraudulent transactions are less than 0.2% of all transactions, but cause significant monetary loss.

    Certain features (e.g., V14, V10) have strong separation between fraud and non-fraud.

    Fraud rates spike during specific hours, suggesting targeted monitoring could be effective.

    Outlier transactions (by amount or feature values) are more likely to be fraudulent.

    Business Impact

    Enhanced fraud detection: Insights can inform rule-based or machine learning models.

    Cost reduction: Time-based and feature-driven monitoring can reduce false positives and investigation workload.

    Actionable recommendations: Focus on high-risk hours and anomalous transactions for real-time alerts.
